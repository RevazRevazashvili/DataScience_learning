```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = [k*k for k in range(10)]
```

```python
plt.plot(y, label=' X^2 ')
plt.title('y = x^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks(x)
plt.yticks(y)
plt.legend()
plt.show()
```

```python
x_ax = np.sin([k*k for k in range(10)])
x_ax
plt.plot(x_ax)
plt.xlabel("X axis")
plt.ylabel('SIN(x)')
plt.show()
```

```python
x = np.arange(10)
plt.plot(np.sin(x))
plt.show()
```
