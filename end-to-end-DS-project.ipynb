```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('VNL2023.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
numeric_cols = df.select_dtypes(include=["int","float"]).columns
corr_mat = df[numeric_cols].corr()
corr_mat
```

```python
sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap=sns.diverging_palette(20,220, as_cmap=True))
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidth=5)
plt.title("correlation matrix heatmap")
plt.show()
```

```python
position_counts = df['Position'].value_counts()
position_counts
```

```python
plt.pie(position_counts, labels=position_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("distribution of positions")
plt.show()
```

```python
avg_attack_by_country = df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False)
```

```python
avg_attack_by_country.sort_values(ascending=False).plot(kind="bar")
plt.title("average attack by country")
plt.xlabel("Country")
plt.ylabel("Average Attack")
plt.show()
```
