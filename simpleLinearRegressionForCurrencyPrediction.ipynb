```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://nbg.gov.ge/monetary-policy/currency'
```

```python
df = pd.read_csv('file.csv')
```

```python
int(df['Date'][0][3:5])
dates = []
for i in range(len(df['Rate'])):
  if df['Date'][i][4:5]=='/':
    dates.append(int(df['Date'][i][3:4]))
  else:
    dates.append(int(df['Date'][i][3:5]))
```

```python
plt.scatter(dates, df['Rate'])
plt.xlabel('Day')
plt.ylabel('EUR cost')
plt.show()
```

```python
poly_1 = np.polyfit(dates,df['Rate'],1)
b1,b0 = poly_1
```

```python
dates = np.array(dates)
y_values = b0  + b1 * dates
```

```python
plt.scatter(dates, df['Rate'])
plt.xlabel('Day')
plt.ylabel('EUR cost')
plt.plot(dates, y_values)
plt.show()
```

```python
poly_1 = np.poly1d(poly_1)
poly_1.c
```

```python
yhat_1 = poly_1(dates)
```

```python
plt.scatter(dates, df['Rate'])
plt.xlabel('Day')
plt.ylabel('EUR cost')
plt.plot(dates, yhat_1)
plt.show()
```

```python
rss = np.power(df['Rate']-yhat_1,2).sum()
rss
```

```python
mss = np.power(df['Rate']-yhat_1,2).mean()
mss
```

```python
tss = np.power(df['Rate']-df['Rate'].mean(),2).sum()
```

```python
# R2
1-(rss/tss)
```
