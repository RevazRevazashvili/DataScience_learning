```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('train.csv')
```

```python
raw_data.head()
```

```python
raw_data.isna().sum()
```

```python
nans = np.where(raw_data.Age.isna())[0]
for i in nans:
  raw_data.Age[i] = raw_data.Age.mean()
```

```python
print(f"{round(raw_data.Age.isna().sum()/raw_data.shape[0] * 100.0, 2)}% missing age data")
```

```python
raw_data['Sex'] = raw_data['Sex'].astype('category')
raw_data['Sex'] = raw_data['Sex'].cat.codes
raw_data['Ticket'] = raw_data['Ticket'].astype('category')
raw_data['Ticket'] = raw_data['Ticket'].cat.codes
raw_data['Cabin'] = raw_data['Cabin'].astype('category')
raw_data['Cabin'] = raw_data['Cabin'].cat.codes
raw_data['Embarked'] = raw_data['Embarked'].astype('category')
raw_data['Embarked'] = raw_data['Embarked'].cat.codes
```

```python
raw_data.head()
```

```python
col = ['Name', 'PassengerId', 'Ticket', 'Survived', 'Cabin']
X = raw_data.drop(columns=col)
X
```

```python
y = raw_data['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state=0).fit(X_train_scaled, y_train)
```

```python
log_reg.predict(X_train_scaled)
```

```python
log_reg.score(X_train_scaled, y_train)
```

```python
log_reg.score(X_test_scaled, y_test)
```

```python
log_reg1 = LogisticRegression(random_state=0,
                              C = 0.01,
                              fit_intercept = True,
                              ).fit(X_train_scaled, y_train)
```

```python
log_reg1.predict(X_train_scaled)
```

```python
log_reg1.score(X_train_scaled, y_train)
```

```python
log_reg1.score(X_test_scaled, y_test)
```
